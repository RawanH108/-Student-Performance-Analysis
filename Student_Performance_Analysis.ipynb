# Student Performance Analysis
This project explores how factors like gender, test preparation, and parental education affect student scores in math, reading, and writing.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.info()
```

```python
# Show descriptive statistics (mean, std, min, max, etc.) for the score columns
```

```python
df.describe()
```

```python
# Plot the distribution of math scores with a KDE curve
```

```python
sns.histplot(df['math score'], kde=True)
plt.title("Math Score Distribution")
plt.show()
```

```python
# Compare math scores between male and female students using a boxplot
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Math Scores by Gender')
plt.show()
```

```python
# Compare reading scores by gender
```

```python
sns.boxplot(x='gender', y='reading score', data=df)
plt.title('Reading Scores by Gender')
plt.show()
```

```python
# Compare writing scores by gender
```

```python
sns.boxplot(x='gender', y='writing score', data=df)
plt.title('Writing Scores by Gender')
plt.show()
```

```python
# Compare math scores between students who completed the test preparation course and those who did not
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.title('Math Scores vs Test Preparation')
plt.show()
```

```python
# Analyze math scores based on parents' education level
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education', y='math score', data=df)
plt.xticks(rotation=45)
plt.title('Math Score vs Parental Education Level')
plt.show()
```

```python
# Show correlation between math, reading, and writing scores
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Scores')
plt.show()
```

```python
# - Female students tend to score higher in reading and writing
# - Test preparation significantly improves performance
# - Strong correlation between reading and writing scores
```
